```python
import json
import pathlib
import matplotlib.pyplot as plt
```

```python
# Path to the percentages file created earlier
results_path = pathlib.Path("Qwen_Qwen3-0.6B_survey_results.json")

with open(results_path) as f:
    percentages = json.load(f)
```

```python
# ── 2. Make an output directory named after the JSON file (without .json) ───────
output_dir = results_path.parent / results_path.stem   # e.g. survey_results/
output_dir.mkdir(exist_ok=True)
```

```python
for idx, (question, dist) in enumerate(percentages.items(), start=1):
    labels  = list(dist.keys())
    values  = [dist[label] for label in labels]

    # One chart per figure (no subplots, per project guidelines)
    plt.figure(figsize=(10, 4))
    plt.barh(labels, values)               # horizontal bar chart
    plt.title(f"Q{idx:02d}: {question}", wrap=True)
    plt.xlabel("Percentage (%)")
    plt.xlim(0, 100)

    # Annotate each bar with its exact percentage
    for y, v in enumerate(values):
        plt.text(v + 0.5, y, f"{v:.2f}%", va="center")

    plt.tight_layout()

    # Show the figure in-notebook (comment out if you only need files)
    plt.show()

    # Save a copy to disk
    out_file = output_dir / f"q{idx:02d}_distribution.png"
    plt.savefig(out_file, dpi=150)
    plt.close()
    print(f"Saved {out_file}")
```
